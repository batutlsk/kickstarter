```python
ham_veri_desc=data.describe()
ham_veri_desc


#kategoriye göre usd_goal_real state=successful
data1=data[data['state']=='successful']


ham_veri_desc=data.describe()
ham_veri_desc


#kategoriye göre usd_goal_real state=successful
data1=data[data['state']=='successful']

data2=data1[['main_category','category','usd_goal_real','backers']]
data2=data2.groupby(['main_category','category']).describe()
data2
```

```python
#backers başına düşen ortalama usd_goal_real (kategori bazında)
backers=data2['backers','mean']
usd_goal=data2['usd_goal_real','mean']

avg_back=usd_goal/backers
avg_back=pd.DataFrame(avg_back)
avg_back.columns=['usd_goal_real']
data1
```

```python
#en yüksek backers başına ortalama usd_goal
avg_back.sort_values('usd_goal_real',ascending=False).iloc[0]
```

```python
import matplotlib.pyplot as plt
cl=data1.iloc[:,[0,14]]
cl

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(cl)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
cl
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(cl)
y_kmeans=pd.DataFrame(y_kmeans)
y_kmeans.columns=['Cluster']
y_kmeans
cl1=cl.reset_index()
clusters=pd.merge(cl1,y_kmeans,left_index=True,right_index=True)
clusters
```

```python
y_kmeans.columns=['cluster']
y_kmeans
cl2 = pd.merge(cl, y_kmeans, left_index=False, right_index=False)
cl2
```
