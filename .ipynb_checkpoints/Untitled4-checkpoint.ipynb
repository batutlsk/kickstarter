```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics.pairwise import cosine_similarity
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.cluster import AgglomerativeClustering
```

```python

data=pd.read_csv("/home/elci/repos/kickstarter/data/ks-projects-201801.csv")
data
```

```python
data["label"]=data.index
data
```

```python
data.dtypes
```

```python
data["launched"]=pd.to_datetime(data["launched"])
data["deadline"]=pd.to_datetime(data["deadline"])
data["time"]=data["deadline"]-data["launched"]
data["time"]=data["time"].dt.days
```

```python
data2=data[["usd_goal_real"
            ,"category","main_category","country","state","usd_pledged_real","time","label"]]
```

```python
data.dtypes
```

```python
series=data2.iloc[20000:22222,:]
```

```python
series
```

```python
lb_make = LabelEncoder()
series["state"] = lb_make.fit_transform(series[["state"]])
series["category"] = lb_make.fit_transform(series[["category"]])
series["main_category"] = lb_make.fit_transform(series["main_category"])
series["country"] = lb_make.fit_transform(series["country"])
```

```python
import matplotlib.pyplot as plt
cl=series
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
   kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
   kmeans.fit(cl)
   wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show(),
```

```python
X=series.as_matrix(columns=series.columns.tolist()[0:])
```

```python
k_means = KMeans(init='k-means++', n_clusters=6, n_init=10)

k_means.fit(X)

labels = k_means.labels_
score = metrics.silhouette_score(X, labels, metric='euclidean')
```

```python
clustering = AgglomerativeClustering(linkage='ward', n_clusters=6)

clustering.fit(X)
kom=clustering.n_leaves_
labels2 = clustering.labels_
score2 = metrics.silhouette_score(X, labels2, metric='euclidean')
```
